#### e) Is there a correlation between the number of bytes transferred from the source to the destination and the number of bytes transferred from the destination to the source? If so, what is the strength of the correlation?

```python
kendall = df["src_bytes"].corr(df["dst_bytes"], method='kendall')
pearson = df["src_bytes"].corr(df["dst_bytes"], method='pearson')
spearman = df["src_bytes"].corr(df["dst_bytes"], method='spearman')

print(f"Kendal: {kendall:2f} \nPearson: {pearson:2f} \nspearman: {spearman:2f}")
```

#### f) Calculate the average amount of data transferred from the source to the destination and from the destination to the source for each (protocol) type of connection.


```python
df["protocol_type"].value_counts()
```

```python
def get_protocol_average(protocol):
    data = df[df["protocol_type"] == protocol]
    return {
        "name": protocol,
        "src_bytes": data["src_bytes"].max() - data["src_bytes"].min(),
        "dst_bytes": data["dst_bytes"].max() - data["dst_bytes"].min()       
        }

def display_protocol(items):
    for i in items:
        print(f"{i['name']}: src_bytes: {i['src_bytes']} and dst_bytes: {i['dst_bytes']}")

tcp = get_protocol_average("tcp")
udp = get_protocol_average("udp")
icmp = get_protocol_average("icmp")

display_protocol([tcp, udp, icmp])
```

#### g) Does the type of connection have an effect on the amount of data transferred

As seen above we can tell that TCP has far more data transfered compared to other transport protocols such as UDP and ICMP. So yes the type of connection has an effect on the amount of data transfered.

#### h) Calculate the distribution of the duration of each type of connection.

```python
# getting the distribution of each type of connection 
def get_duration_distribution(protocol):
    data = df[df["protocol_type"] == protocol]
    return data["duration"].describe()

for i in ["tcp", "udp", "icmp"]:
    print(i.upper())
    print(get_duration_distribution(i))
```

#### i) Is there a correlation between the duration of a connection and the amount of data transferred?

```python
pearson_dst = df["duration"].corr(df["dst_bytes"])
pearson_src = df["duration"].corr(df["src_bytes"])

kendall_dst = df["duration"].corr(df["dst_bytes"], method='kendall')
kendall_src = df["duration"].corr(df["src_bytes"], method='kendall')

spearman_dst = df["duration"].corr(df["dst_bytes"], method='spearman')
spearman_src = df["duration"].corr(df["src_bytes"], method='spearman')

print(f"pearson correlation of dst {pearson_dst:2f} and src {pearson_src:2f}")
print(f"Kendall correlation of dst {kendall_dst:2f} and src {kendall_src:2f}")
print(f"Spearman correlation of dst {spearman_dst:2f} and src {spearman_src:2f}")
```

There seems to be no correlation between the duration of a connection and amount of data transfered. Which is suprising because you would think the more bytes transfered would lead to a longer connection duration.

#### j) Calculate the average number of bytes transferred per second for each type of connection

```python
# formula: (max bytes - min byes)/ mean duration 

def average_num_bytes(protocal):
    data = df[df["protocol_type"] == protocal]
    return {
        'src_dst': (data["src_bytes"].max() - data["src_bytes"].min()) / data["duration"].mean(),
        'dst_bytes': (data["dst_bytes"].max() - data["dst_bytes"].min()) / data["duration"].mean(),
    }

for i in ["tcp", "udp"]:
    print(i.upper())
    avgs = average_num_bytes(i)
    print(f"Average per second from source to destination: {avgs['src_dst']}")
    print(f"Average per second from destination to source: {avgs['dst_bytes']}")
```
