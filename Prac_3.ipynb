#### f) Calculate the average amount of data transferred from the source to the destination and from the destination to the source for each (protocol) type of connection.


```python
df["protocol_type"].value_counts()
```

```python
def get_protocol_average(protocol):
    data = df[df["protocol_type"] == protocol]
    return {
        "name": protocol,
        "src_bytes": data["src_bytes"].max() - data["src_bytes"].min(),
        "dst_bytes": data["dst_bytes"].max() - data["dst_bytes"].min()       
        }

def display_protocol(items):
    for i in items:
        print(f"{i['name']}: src_bytes: {i['src_bytes']} and dst_bytes: {i['dst_bytes']}")

tcp = get_protocol_average("tcp")
udp = get_protocol_average("udp")
icmp = get_protocol_average("icmp")

display_protocol([tcp, udp, icmp])
```

#### j) Calculate the average number of bytes transferred per second for each type of connection

```python
# formula: (max bytes - min byes)/ mean duration 

def average_num_bytes(protocal):
    data = df[df["protocol_type"] == protocal]
    return {
        'src_dst': (data["src_bytes"].max() - data["src_bytes"].min()) / data["duration"].mean(),
        'dst_bytes': (data["dst_bytes"].max() - data["dst_bytes"].min()) / data["duration"].mean(),
    }

for i in ["tcp", "udp"]:
    print(i.upper())
    avgs = average_num_bytes(i)
    print(f"Average per second from source to destination: {avgs['src_dst']}")
    print(f"Average per second from destination to source: {avgs['dst_bytes']}")
```

#### Select two groups of top 10 relevant features: i) using the correlation heatmap of 2b) and ii) using regression analysis.


```python
top_10_heatmap_features = ["flag", "logged_in", "same_srv_rate", 
                           "dst_host_srv_count", "dst_host_same_srv_rate",
                           "protocol_type", "dst_bytes", "src_bytes",
                           "srv_diff_host_rate", "is_guest_login" 
                           ]
```

```python
# i will be using logistic regression to using binary_labels are the target
# i will use it with every feature and the ones with the highest scores will be added to the feature list
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

ml_encode = df_encode.copy()
reg = LogisticRegression()

y = ml_encode["binary_labels"]
x = ml_encode.drop(labels=["binary_labels", "labels"], axis=1)

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

results = {}
for i in x_train.columns:
    curr_train_X = np.column_stack((np.ones(len(x_train[i])), x_train[i]))
    curr_test_X = np.column_stack((np.ones(len(x_test[i])), x_test[i]))
    reg.fit(curr_train_X, y_train)
    results[i] = reg.score(curr_test_X, y_test)

```

```python
#sorting the results
sorted_results = sorted(results.items(), key=lambda item: item[1], reverse=True)

top_10_regression = []

for index, value in enumerate(sorted_results):
    if index == 10:
        break 
    top_10_regression.append(value[0])

print(top_10_regression)
```

####  Prepare a tree-based machine learning model by preprocessing the 70:30 train-test split data, measuring appropriate performance metrics but with a focus on maximizing macro f1-score,and contrasting the two groups of top 10 modelled features

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, accuracy_score


df_tree = df_encode.copy()

# Models
tree_multi_reg = DecisionTreeClassifier()
tree_multi_corr = DecisionTreeClassifier()
tree_binary_reg = DecisionTreeClassifier()
tree_binary_corr = DecisionTreeClassifier()

X = df_tree.drop(labels=["binary_labels", "labels"], axis=1)

mult_y = df_tree["labels"]
binary_y = df_tree["binary_labels"]

mulit_X_train, mulit_X_test, mulit_y_train, mulit_y_test = train_test_split(X, mult_y, test_size=0.3, random_state=42)
binary_X_train, binary_X_test, binary_y_train, binary_y_test = train_test_split(X, binary_y, test_size=0.3, random_state=42)


multi_class_reg = tree_multi_reg.fit(mulit_X_train[top_10_regression], mulit_y_train)
multi_class_corr = tree_multi_corr.fit(mulit_X_train[top_10_heatmap_features], mulit_y_train)

binary_reg = tree_binary_reg.fit(binary_X_train[top_10_regression], binary_y_train)
binary_corr = tree_binary_corr.fit(binary_X_train[top_10_heatmap_features], binary_y_train)

```

```python
#results
def output_results(name, models, score_Func,  XTest, Y_test):
    print(name)


    predict_multiclass_reg = models[0].predict(XTest[0])
    predict_multiclass_corr = models[1].predict(XTest[1])

    print(f"multiclass score for regression: {score_Func(Y_test[0], predict_multiclass_reg)}")
    print(f"multiclass score for corr: {score_Func(Y_test[0], predict_multiclass_corr)}")

    predict_binary_reg = models[2].predict(XTest[2])
    predict_binary_corr = models[3].predict(XTest[3])
    print(f"Binary score for regression: {score_Func(Y_test[1], predict_binary_reg)}")
    print(f"Binary score for corr: {score_Func( Y_test[1], predict_binary_corr)}")


models = [multi_class_reg, multi_class_corr, binary_reg, binary_corr]
X_Test = [mulit_X_test[top_10_regression], mulit_X_test[top_10_heatmap_features], binary_X_test[top_10_regression], binary_X_test[top_10_heatmap_features]]
Y_test = [mulit_y_test, binary_y_test]

output_results("Accuaracy", models, accuracy_score, X_Test, Y_test)
f1_weight = lambda y_true, y_pred: f1_score(y_true, y_pred, average="macro")

output_results("F1_score Macro", models, f1_weight, X_Test, Y_test)
```

#### Perform appropriate hypothesis tests that answer the research questions as a basis for Task 2.

### Task 2: Results Analysis and Report

What is the most prevalent attack in the NSL-KDD Network Intrusion Detection dataset?  
● H0: All types of attacks occur with equal frequency in the dataset.  
● H1: There is a difference in the frequency of different types of attacks in the dataset.

```python
from scipy.stats import chi2, chi2_contingency


# contiengency table
crosstab = pd.crosstab(index=df_bin.index, columns=df_bin['binary_labels'])
crosstab.columns = ['Normal', 'Attack'] 

# performing the test 
chi2_stat, p_value, dof, expected = chi2_contingency(crosstab)

# Output the Chi-Square Test results
print(f'Chi-Square Statistic: {chi2_stat}')
print(f'p-value: {p_value}')
print(f'Degrees of Freedom: {dof}')

# Calculate the critical value for the Chi-Square Distribution
alpha = 0.05  # Significance level
critical_value = chi2.ppf(1 - alpha, df=dof)
print(f'Critical Value at alpha={alpha} and df={dof}: {critical_value:.2f}')

# Compare the Chi-Square Statistic with the Critical Value
if chi2_stat > critical_value:
    print("Reject the null hypothesis: significant association between normal attacks and the different types of attacks.")
else:
    print("Fail to reject the null hypothesis: no significant association between normal attacks and the different types of attacks.")
```
